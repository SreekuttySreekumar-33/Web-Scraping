```python
data={"product_page_url":[],
      "title" : [], 
      'universal_product_code' : [],
      'price_including_tax': [],
      'price_excluding_tax' : [],
      'number_available' : [],
      'category' : [],
      'review_rating' : []
      }
```

```python
baseurl = "https://books.toscrape.com/"
for li in li_list:
  href_split= li.a["href"].split("/")[3:]
  url = baseurl + "catalogue/" + "/".join(href_split)
  response = requests.get(url)
  soup = BeautifulSoup(response.content)
  data["product_page_url"].append(url)
  title = soup.find("h1").text.strip()
  data["title"].append(title)
  rows = soup.find("table").find_all("tr")
  UPC = rows[0].find("td").text.strip()
  data['universal_product_code'].append(UPC)
  data["category"].append("Travel")
  PIT = rows[3].find("td").text.strip()
  data['price_including_tax'].append(PIT)
  PET = rows[2].find("td").text.strip()
  data['price_excluding_tax'].append(PET)
  num= rows[5].find("td").text.strip()
  data['number_available'].append(num)
  rev= soup.find_all("p")[2]["class"][1]
  data['review_rating'].append(rev)
```

```python
data
```

```python
import pandas as pd
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("scraping_output.csv",index=False)
```
